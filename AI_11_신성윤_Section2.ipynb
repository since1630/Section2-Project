```python
# 평가지표 함수
def get_eval(y_val,y_pred,y_pred_proba=None):
  
  # r2 = r2_score(y_val, y_pred)
  # mae = mean_absolute_error(y_val, y_pred)
  # print('R2스코어: {0:.2f}, MAE스코어: {1:.2f}, 기준모델MAE스코어:{2:.2f}'.format(r2,mae,base_mae))
  
  confusion = confusion_matrix(y_val, y_pred)
  print('오차행렬')
  print(confusion)

  accuracy = accuracy_score(y_val, y_pred)
  precision = precision_score(y_val, y_pred)
  recall = recall_score(y_val, y_pred)
  f1 = f1_score(y_val, y_pred)
  roc_auc = roc_auc_score(y_val, y_pred_proba)

  print('정확도:{0:.2f}, 정밀도:{1:.2f}, 재현율:{2:.2f}, F1스코어:{3:.2f}, AUC스코어:{4:.2f}'.format(accuracy,precision,recall,f1,roc_auc))
```

```python
# 모델 및 학습/검증 데이터를 받아 학습,예측,평가 수행하는 함수
def model_and_eval(model,X_train,X_val,y_train,y_val):

  model.fit(X_train, y_train)
  y_pred = model.predict(X_val)
  y_pred_proba = model.predict_proba(X_val)[:,1]
  get_eval(y_val, y_pred, y_pred_proba)
  
  
```

# 분류로 풀어보자.
##( 회귀로 풀 경우 highcardinality 가 매우 많으므로 저것들을 전부 처리하기는 힘들다. 아무래도 분류로 풀어야 할 것 같긴한데... 분류로 풀어야 할 경우 타겟을 이진분류로 바꾸던가 다중분류로 풀어야 하는데 다중분류는 배우지 않았고 타겟을 이진분류로 바꾸는것은 애매모호한 부분이 있다( 평점이 5/4/3/2/1 이 있는데 평점의 좋고 나쁜 기준이 애매모호함)  -> 회귀모델을 테스트해봤으나 성능이 너무 떨어져 분류로 접근하기로했다.

```python
# 랜덤포레스트 분류모델 생성
pipe_rf = make_pipeline(OrdinalEncoder(),
                     RandomForestClassifier(random_state=2,class_weight = 'balanced')
                     )

model_and_eval(pipe_rf,X_train,X_val,y_train,y_val)
print('기준모델 정확도:{0:.2f}, 정밀도:{1:.2f}, 재현율:{2:.2f}, F1스코어:{3:.2f}'.format(base_accuracy, base_precision, base_recall, base_f1))
```

```python
# 랜덤포레스트 분류모델 생성 ( 타겟 인코딩 )
pipe_rf = make_pipeline(TargetEncoder(),
                     RandomForestClassifier(random_state=2,class_weight = 'balanced')
                     )

model_and_eval(pipe_rf,X_train,X_val,y_train,y_val)
print('기준모델 정확도:{0:.2f}, 정밀도:{1:.2f}, 재현율:{2:.2f}, F1스코어:{3:.2f}'.format(base_accuracy, base_precision, base_recall, base_f1))
```

```python
# XG부스팅 모델 생성
pipe_xgb = make_pipeline(OrdinalEncoder(),
                     XGBClassifier(random_state=2, scale_pos_weight = 3.0766)
                     )

model_and_eval(pipe_xgb,X_train,X_val,y_train,y_val)
print('기준모델 정확도:{0:.2f}, 정밀도:{1:.2f}, 재현율:{2:.2f}, F1스코어:{3:.2f}'.format(base_accuracy, base_precision, base_recall, base_f1))
```

```python
# XG부스팅 모델 생성 ( 타겟 인코딩 )
pipe_xgb = make_pipeline(TargetEncoder(),
                     XGBClassifier(random_state=2, scale_pos_weight = 3.0766)
                     )

model_and_eval(pipe_xgb,X_train,X_val,y_train,y_val)
print('기준모델 정확도:{0:.2f}, 정밀도:{1:.2f}, 재현율:{2:.2f}, F1스코어:{3:.2f}'.format(base_accuracy, base_precision, base_recall, base_f1))
```
